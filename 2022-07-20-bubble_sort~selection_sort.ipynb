# 선택정렬 알고리듬 

```python
# 리스트 내 최댓값 찾는 알고리듬 사고실험 결과 테스트 
import numpy as np 
import matplotlib.pyplot as plt 

x = list(np.random.sample(100)) ; print(type(x))
plt.bar(range(100), x)
```

```python
# 최댓값 찾는 알고리듬 
def find_maximum(lis) : 
    max = lis[0] # 최댓값 기본 설정 : 0번 인덱스 
    for i in range(1, len(lis)) : 
        if max < lis[i] : 
            max = lis[i]
    max_index = lis.index(max)
    return max, max_index
find_maximum(x)
```

```python
# 선택정렬 알고리즘 구현 
def selection_sort(x) : 
    li = len(x)
    for i in range(len(x)-1) : # n차 패스 진행
        unsorted_part = x[:li] # 정렬 안 된 부분이 있다
        max, max_index = find_maximum(unsorted_part) # 정렬 안 된 부분에서 최댓값 찾는다
        x[max_index], x[li-1] = x[li-1], x[max_index] # 최댓값과 정렬 안 된 부분 가장 오른쪽 값 교환한다. 
        li -= 1 # 정렬 안 된 부분 크기 하나 줄인다. 
    return x
```

```python
r = selection_sort(x)
plt.bar(range(100), r)
```

# 선택 정렬 
### 1. 정렬 안 된 부분은 오른쪽 끝에서부터 하나씩 줄어든다. 
### 2. 정렬 안 된 부분에서 최댓값 찾는다. 
### 3. 정렬 안 된 부분 오른쪽 끝과 최댓값 교환한다. 

```python
# 책에 있는 방식으로 선택정렬 구현 
def selection_sort2(lis) : 
    for i in range(len(lis)-1, 0, -1) : # i = 오른쪽 끝 위치 

        # 최댓값 위치 찾기 
        max_index = 0 
        for y in range(i+1) : 
            if lis[max_index] < lis[y] : 
                max_index = y 
        
        # 최댓값과 오른쪽 끝 교환
        lis[i], lis[max_index] = lis[max_index], lis[i]
    return lis
```

```python
plt.bar(range(100), selection_sort2(r))
```

```python
list4 = [70, 15, 25, 19, 34, 44]
result = selection_sort2(list4)
print(result)
```

```python
# 복습 
# 버블정렬 

def bubble_sort(x) : 
    for end in range(len(x)-1, 0, -1) : 
        for i in range(0,end) : 
            if x[i] > x[i+1] : 
                x[i], x[i+1] = x[i+1], x[i]
    return x
```

```python
x = [25, 21, 22, 24, 23, 27, 26]
bubble_sort(x)
```

```python
# 삽입 정렬 복습 
def insert_sort(x) : 
    for n in range(1, len(x)) : 
        j = n # 정렬 해야 할 가장 왼쪽 위치
        temp = x[j]
        while (j-1 >= 0) and (temp<x[j-1]) : 
            x[j] = x[j-1]
            j -= 1
        x[j] = temp 
    return x
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = [25, 26, 22, 24, 27, 23, 21]
insert_sort(x)
x2 = list(np.random.sample(100))
plt.bar(range(100), insert_sort(x2))
```

```python
# 병합 정렬 
def merge_sort(x) : 
    # 분리 
    if len(x) > 1 : # 분리 조건 
        separate_criterion = len(x)//2 
        left = x[:separate_criterion]
        right = x[separate_criterion:]

        merge_sort(left)
        merge_sort(right)

        # 병합 
        a = 0 
        b = 0 
        c = 0 

        while (a < len(left)) and (b < len(right)) : 
            if left[a] > right[b] : 
                x[c] = right[b]
                b += 1
            else : 
                x[c] = left[a]
                a += 1 
            c += 1 

        while (a < len(left)) : 
            x[c] = left[a]
            a += 1
            c += 1
        
        while (b < len(right)) : 
            x[c] = right[b]
            b += 1
            c += 1

    return x 
```

```python
list3 = [44, 16, 83, 7, 67, 21, 34, 45, 10]
merge_sort(list3)
```

```python
# 셸 정렬 
def shell_sort(x) : 
    distance = len(x)//2 # 거리 지정 
    
    while distance > 0 : 
        # 부분 리스트 정렬 알고리듬
        for i in range(distance, len(x)) : 
            j = i 
            while (j-distance >= 0) and (x[j] < x[j-distance]) : 
                x[j], x[j-distance] = x[j-distance], x[j]
                j = j - distance 
            
        distance = distance // 2
    
    return x 
```

```python
x = list(np.random.sample(100))
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(range(100), x)
plt.subplot(1,2,2)
plt.bar(range(100), shell_sort(x))
plt.show()
```

```python
# 선택정렬 복습 
def selection_sort(x) : 
    for r in range(len(x)-1, 0, -1) : # 가장 오른쪽 요소의 인덱스 (1씩 감소)

        # 가장 큰 값 찾기 
        max_index = 0 
        for i in range(1, r+1) : 
            if x[max_index] < x[i] : 
                max_index = i 
                
        # 교환 
        x[max_index], x[r] = x[r], x[max_index]
    return x 
```

```python
x = np.random.sample(500) 
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.bar(range(500), x)

plt.subplot(1,2,2)
plt.bar(range(500), selection_sort(x))
```
