# insertion algorithm

```python
# 삽입 알고리듬 
def inssertion_algorithm(contents) : 
    for i in range(1, len(contents)) : # i = 정렬안된 부분의 현재위치 포인터 
        while (i != 0) : 
            if contents[i-1] > contents[i] : 
                contents[i-1], contents[i] = contents[i], contents[i-1] # 서로 교환
                i -= 1
            else : break 
    return contents 
```

```python
# 테스트 1
list1 = [25, 26, 22, 24, 27, 23, 21] 
inssertion_algorithm(list1)
```

```python
# 테스트 2
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
list2 = list(np.random.sample(100))
result = inssertion_algorithm(list2)

plt.figure(figsize=(100,50))
plt.bar(range(100), result)
plt.title('insertion algorithm works?')
plt.ylabel('values')
plt.xlabel('result order')
plt.show()
```

```python
# 삽입 정렬 -2 (책이 제시하는 방식 이해 후 작성)
def insertion_sort(contents) : 
    for i in range(1, len(contents)) : 
        j = i - 1 # 이전 위치 순서 
        present_location = contents[i] # 정렬할 값 

        while (j >= 0 and contents[j] > present_location) : # j >= 0 은 기본 조건. contents[j] > present location 은 정렬할 위치에 있는 값과 그 전 값 크기 비교. 
            contents[j+1] = contents[j] # 조건 만족하면; 이전 위치에 있던 값이 뒤로 한 칸 이동. 빈 자리에 '정렬할 값'이 삽입 되었다고 개념적으로 가정. 
            j -= 1 
        # 이동이 끝난 후 '정렬할 값' 물리적으로 삽입
        contents[j+1] = present_location
        
    return contents 
```

```python
# 알고리듬 테스트 
result2 = insertion_sort(list2)
plt.figure(figsize=(100, 50))
plt.bar(range(100), result2)
plt.show()
```

```python
# 다시 
def insertion_sort2(cont) : 
    for i in range(1, len(cont)) : # 정렬 안 된 부분에서 현재위치 포인터 i 가 1~len(list)-1 까지 순차 이동한다. 
        mustbesorted = cont[i] # 정렬해야 할 값 i에 대해서
        j = i - 1 # i의 이전값 j 정의
        while (j >= 0) and (cont[j] > mustbesorted) : 
            cont[j+1] = cont[j] # 만약 j가 더 크면 i 위치로 j가 이동한다. 
            j -= 1 # j를 1 줄인다
        cont[j+1] = mustbesorted # 리스트 j+1 위치에 i를 물리적 삽입한다.
    return cont # 정렬 결과 반환
```

# merge sort 

```python
# 병합정렬 실습 (분리, 병합)
def mergesort(cont) : 
    # 분리 
    if len(cont) > 1 :  
        mid_point = len(cont)//2 # 분리 하기 위한 중심점 설정 
        # 왼쪽 분리 
        left = cont[:mid_point] 
        # 오른쪽 분리 
        right = cont[mid_point:]

        # 분리된 왼쪽과 오른쪽에서, 크기 1짜리 요소 하나만 남을 때 까지 다시 분리. 
        mergesort(left) 
        mergesort(right)
        #------------------------------------
        # 정렬된 왼쪽과 오른쪽 리스트 병합

        # 인덱스 값 설정 
        a = 0 # 왼쪽 리스트 인덱스 
        b = 0 # 오른쪽 리스트 인덱스 
        c = 0 # 원본 리스트 인덱스 

        while (a < len(left) and b < len(right)) : # 만약 왼쪽과 오른쪽 리스트 모두 원본 리스트에 안 넣은 요소 남아있는 경우 
            if left[a] < right[b] : 
                cont[c] = left[a]
                a += 1 
            else : 
                cont[c] = right[b]
                b += 1
            c += 1 
        while (a < len(left)) : # 만약 오른쪽 리스트가 원본 리스트에 정렬할 요소 다 떨어진 경우; 왼쪽 리스트만 남은 경우 
            cont[c] = left[a]
            a += 1 
            c += 1
        while (b < len(right)) : # 만약 왼쪽 리스트가 원본 리스트에 정렬할 요소 다 떨어진 경우; 오른쪽 리스트만 남은 경우 
            cont[c] = right[b]
            b += 1
            c += 1
        # <정렬 끝>
    return cont
```

```python
# 알고리듬 실행
example = np.random.sample(100)
plt.figure(figsize=(10,5))
plt.bar(range(100), example) # 정렬되지 않은 상태다. 
```

```python
example = np.random.sample(100)
result = mergesort(example)
plt.figure(figsize=(10,5))
plt.bar(range(100), result)
plt.show() # 정렬되었다. 
```

# shell sort 

```python
# 셸 정렬 
def shellsort(cont) : 
    dist = len(cont) // 2 # 요소 간 거리 지정 

    while (dist > 0) : # distance 가 0이다 = 정렬이 완료됬다 는 의미다. 
        # 하위 부분 리스트 정렬하는 알고리듬 
        for i in range(dist, len(cont)): 
            j = i # j 는 하위 부분 리스트 정렬 안 된 부분 가장 왼쪽 원소 의미한다. 
            while (j-dist >= 0) and (cont[j] < cont[j-dist]) : 
                cont[j], cont[j-dist] = cont[j-dist], cont[j] # 정렬 된 부분 가장 오른쪽 원소와 크기 비교 후 조건 충족하면 서로 교환
                j = j - dist # j 위치 바뀌었기 때문에 j = j-distance 로 j 변수에 값 새로 할당해 준다. 
        
        dist = dist // 2 # 보통 셸 알고리듬에서 요소 간 거리는 절반씩 줄여나간다. 

    return cont 
```

```python
list3 = [26, 17, 20, 11, 23, 21, 13, 18, 24, 14, 12, 22, 16, 15, 19, 25]
shellsort(list3)
```
